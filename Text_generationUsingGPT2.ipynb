```python
import tensorflow as tf
import transformers
from numpy import random
from transformers import TFGPT2LMHeadModel,GPT2Tokenizer,GPT2Config
```

```python
print('tensorflow version>>>',tf.__version__)
print('hungginface transformers version>>>>',transformers.__version__
      )
```

#Model Setup

```python
model_name = 'gpt2-medium'
config = GPT2Config.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2LMHeadModel.from_pretrained(model_name,config=config)
```

#Generate text

```python
#encode context the generation is conditioned on
input_ids = tokenizer.encode('hey what are',return_tensors='tf')
input_ids
```

```python
# As it is word level model not character level model
output = model.generate(input_ids,max_length=15)
output = tokenizer.decode(output[0],skip_special_tokens=True)
print('>'*100)
print('output:- ',output)
```
